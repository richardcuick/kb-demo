```matlab
tx = linspace (-8, 8, 41);
ty = tx;
[xx, yy] = meshgrid (tx, ty);
r = sqrt (xx .^ 2 + yy .^ 2) + eps;
tz = sin (r) ./ r;
```

```matlab
%plot inline
mesh(tx, ty, tz);
```

```matlab
%plot native
mesh(tx, ty, tz);
```
